# Face Detection

```python
import cv2
xml = r'C:\Users\Haier\Desktop\kah\face.xml' # harcascade file (library to detect face) 
path = r'C:\Users\Haier\Desktop\kah\faces.jpg'  
img = cv2.imread(path)
img
```

```python
face_cascade = cv2.CascadeClassifier(xml)
```

```python
faces = face_cascade.detectMultiScale(img,1.2,5)
```

```python
faces
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 3)
cv2.imshow("Faces",img) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Eyes Detection

```python
import cv2
xml = r'C:\Users\Haier\Desktop\kah\face.xml' # harcascade file (library to detect face) 
eyes_xml =  r'C:\Users\Haier\Desktop\kah\eye.xml'

path = r'C:\Users\Haier\Desktop\kah\faces.jpg'  
img = cv2.imread(path)

face_cascade = cv2.CascadeClassifier(xml)
eye_cascade = cv2.CascadeClassifier(eyes_xml)
faces = face_cascade.detectMultiScale(img,1.2,5)


for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 3)
    my_face = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(my_face)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(my_face, (ex,ey), (ex+ew, ey+eh), (0,255,0), 2)
cv2.imshow("Faces",img) 
cv2.waitKey(0)
cv2.destroyAllWindows()
        
```

# Smile Detection

```python
import cv2
xml = r'C:\Users\Haier\Desktop\kah\face.xml' # harcascade file (library to detect face) 
smile_xml =  r'C:\Users\Haier\Desktop\kah\smile.xml'

path = r'C:\Users\Haier\Desktop\kah\smiles.jpg'  
img = cv2.imread(path)

face_cascade = cv2.CascadeClassifier(xml)
smile_cascade = cv2.CascadeClassifier(smile_xml)
faces = face_cascade.detectMultiScale(img,1.2,5)


for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 3)
    my_face = img[y:y+h, x:x+w]
    smile = smile_cascade.detectMultiScale(my_face,1.8, 15)
    for (sx,sy,sw,sh) in smile:
        cv2.rectangle(my_face, (sx,sy), (sx+sw, sy+sh), (0,255,0), 2)
cv2.imshow("Faces",img) 
cv2.waitKey(0)
cv2.destroyAllWindows()
        
```
