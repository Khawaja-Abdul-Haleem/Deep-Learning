```
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Data Set
training_data =np.array([[0,0], [0,1], [1,0], [1,1]])
training_data
```

```
target_data = np.array([[0], [1], [1], [0]])
target_data
```

```
# Build/Create Model model
model = tf.keras.models.Sequential() # Blank Neural Network
```

```
# add layers
model.add(tf.keras.layers.Dense(4, input_dim =2, activation = 'relu')) # 4 neurons, Input Layer
model.add(tf.keras.layers.Dense(1, activation ='sigmoid'))
 
```

```
# Configuring Model
# define loss function and optimizer
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['binary_accuracy'])
```

```
# now we are ready to apply model and train
trained_model = model.fit(training_data, target_data, epochs = 1000)
```

```
ypred = model.predict(training_data)
```

```
ypred
```

```
ypred.round()
```

```
# Plotting Loss Vs Epoch
loss_curve = trained_model.history['loss']
plt.title("Loss Vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.plot(loss_curve, color = 'red')
plt.show()
```

```
# Plotting Loss Vs Epoch
acc_curve = trained_model.history['binary_accuracy']
plt.title("Accuracy Vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.plot(acc_curve, color = 'blue')
plt.show()
```
